## 1. Conectarse al servidor

Cambia `SERVER_IP` por la IP pública de tu VM (la misma donde corre el servidor de Socket.IO).

```
from socketIO_client import SocketIO

SERVER_IP = "TU-IP-SERVIDOR"   # ej: "34.136.249.24"
SERVER_PORT = 5001

print("Conectando...")
try:
    socketIO = SocketIO(SERVER_IP, SERVER_PORT)
    print("✅ Conectado al servidor.")
except Exception as e:
    print(f"❌ Error al conectar: {e}")
```

## 2. Funciones de control

Son exactamente la misma lógica que el cliente de escritorio original: arman un mensaje `{'motor': ..., 'command': ...}` y lo emiten por el evento `ctrl_from_python`.

```
def increase(motor):
    message = {"motor": motor, "command": "+"}
    socketIO.emit("ctrl_from_python", message)
    return f"Enviado: {message}"


def decrease(motor):
    message = {"motor": motor, "command": "-"}
    socketIO.emit("ctrl_from_python", message)
    return f"Enviado: {message}"
```

## 3. Interfaz gráfica con Gradio

Reemplaza la ventana de Tkinter (que no funciona en Colab) por una interfaz web interactiva.

```
import gradio as gr

with gr.Blocks() as app:
    gr.Markdown("### 🕹️ CONTROL — Cliente 1 (Botones)")

    motor_menu = gr.Dropdown(
        choices=["1", "2", "3", "4", "5"],
        value="1",
        label="Seleccionar Motor",
    )

    with gr.Row():
        btn_increase = gr.Button("➕", size="lg")
        btn_decrease = gr.Button("➖", size="lg")

    salida = gr.Textbox(label="Consola", interactive=False)

    btn_increase.click(fn=increase, inputs=motor_menu, outputs=salida)
    btn_decrease.click(fn=decrease, inputs=motor_menu, outputs=salida)

app.launch(inline=True, share=True)
```
